```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Analise_mercado_imobiliário.xlsx', sheet_name='0. BASE DE TUDO')
```

```python
filtered_data = data[(data['ZONA'] == 'CENTRO')]
# Definir os intervalos (bins)
bin = [0, 250000, 500000, 750000, 1000000, 1250000, 1500000, 1750000, 2000000]

# Criar o histograma
ax = sns.histplot(x='VALOR DA CASA', data=filtered_data, bins=bin, kde=False, stat="percent")

plt.xlabel('Valor da casa')
plt.ylim(0, 100)
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

plt.ylabel('Frequência')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Leitura dos dados
data = pd.read_excel('Analise_mercado_imobiliário.xlsx', sheet_name='0. BASE DE TUDO')
data = data[data['ZONA'] == 'NORTE']

# Defina os valores de x para o dotplot
x_values = data['NÚMERO DE BANHEIROS']

# Crie o gráfico de pontos usando a função dotplot
def dotplot(input_x, **args):
    # Contar quantas vezes cada valor ocorre
    unique_values, counts = np.unique(input_x, return_counts=True)

    # Converter a entrada 1D em uma matriz 2D
    scatter_x = []  # valores de x
    scatter_y = []  # valores de y correspondentes
    for idx, value in enumerate(unique_values):
        for counter in range(1, counts[idx] + 1):
            scatter_x.append(value)
            scatter_y.append(counter)

    # Desenhar o gráfico de pontos usando scatter()
    plt.scatter(scatter_x, scatter_y, **args)

# Chame a função dotplot com os valores de x corretos
dotplot(input_x=x_values, color='blue')  # Mude a cor dos pontos para vermelho

# Defina os limites do eixo x com base nos valores de x
plt.xlim(min(x_values) - 1, max(x_values) + 1)
plt.xlabel('Número de Banheiros')
plt.ylabel('Frequência')

# Adicione um título ao gráfico
plt.title('Número de Banheiros na Zona Norte')

# Exiba o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Leitura dos dados
data = pd.read_excel('Analise_mercado_imobiliário.xlsx', sheet_name='0. BASE DE TUDO')

# Filtrar os dados por casas com 3 quartos
data_3_quartos = data[data['NÚMERO DE QUARTOS'] == 3]

# Contar o número de casas com 3 quartos em cada região
casas_3_quartos_por_regiao = data_3_quartos['ZONA'].value_counts()

# Calcular a porcentagem
porcentagem_por_regiao = (casas_3_quartos_por_regiao / casas_3_quartos_por_regiao.sum()) * 100

# Rótulos das fatias do gráfico de pizza
labels = porcentagem_por_regiao.index

# Valores das fatias do gráfico de pizza
sizes = porcentagem_por_regiao.values

# Criar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Porcentagem de Casas com 3 Quartos por Região')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Leitura dos dados
data = pd.read_excel('Analise_mercado_imobiliário.xlsx', sheet_name='0. BASE DE TUDO')

# Aplicar os filtros
filtered_data = data[(data['NÚMERO DE QUARTOS'] >= 3) & (data['NÚMERO DE BANHEIROS'] >= 2) & (data['VALOR DA CASA'] <= 350000) & (data['TEM EDÍCULA?'] == 'Sim')]

# Verificar se o DataFrame está vazio
if not filtered_data.empty:
    # Contagem de imóveis por zona
    contagem_por_zona = filtered_data['ZONA'].value_counts(normalize=True) * 100  # Calcula as porcentagens

    # Criar um gráfico de barras
    plt.figure(figsize=(10, 6))
    contagem_por_zona.plot(kind='bar', color='skyblue')
    plt.title('Porcentagem de Imóveis por Zona com as Características Desejadas')
    plt.xlabel('Zona')
    plt.ylabel('Porcentagem (%)')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Nenhum imóvel corresponde aos critérios especificados.")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

data = pd.read_excel('Analise_mercado_imobiliário.xlsx', sheet_name='BASE DE DADOS (NORTE)')

# Separe as features (X) e o target (Y)
X = data[['VALOR DA CASA']]  # Usando colchetes para criar uma lista de colunas
Y = data['TAMANHO DO TERRENO']  # Uma única coluna como alvo

# Crie um modelo de regressão linear
model = LinearRegression()
model.fit(X, Y)

# Faça previsões com base no modelo
predicted_Y = model.predict(X)

# Calcule os resíduos (diferença entre os valores observados e previstos)
residuals = Y - predicted_Y

# Obtenha os coeficientes da regressão
a = model.coef_[0]
b = model.intercept_

# Calcule o coeficiente de determinação (r-squared)
r_squared = r2_score(Y, predicted_Y)

# Crie um gráfico de regressão linear
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X, Y, label='Valores Observados')
plt.plot(X, predicted_Y, color='red', label=f'Regressão Linear: Y = {a:.9f}X + {b:.2f}\nR² = {r_squared:.2f}')
plt.xlabel("Variável X: tamanho m^2")
plt.ylabel("Variável Y:")
plt.title("Gráfico de Regressão Linear (NORTE)")
plt.legend()

# Crie um gráfico de resíduos
plt.subplot(1, 2, 2)
plt.scatter(X, residuals, c='r')
plt.axhline(y=0, color='b', linestyle='--')
plt.xlabel("Variável X")
plt.ylabel("Resíduos")
plt.title("Gráfico de Resíduos")

# Mostre os gráficos
plt.tight_layout()
plt.show()
```
